### Univariate Analysis for Continuous variables :

```python
# Set the dark background style
plt.style.use('dark_background')

# Set the figure size
plt.figure(figsize=(6, 4))
# Create the kernel density estimate plot with seaborn
sns.kdeplot(data=df['Sales'], fill=True, color='blue', alpha=0.3, linewidth=3)

# Add a vertical line for the median
plt.axvline(df['Sales'].median(), color='red', linestyle='-', linewidth=2)
plt.grid(True)
# Set the x-axis and y-axis labels
plt.xlabel('Sales')
plt.ylabel('Density')

# Add a title to the plot
plt.title('Sales Distribution')

# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(6, 4))
# Create the kernel density estimate plot with seaborn
sns.kdeplot(data=df['Profit'], fill=True, color='blue', alpha=0.3, linewidth=2)

# Add a vertical line for the median
plt.axvline(df['Profit'].median(), color='red', linestyle='-', linewidth=2)
# Set the dark background style
plt.style.use('dark_background')
plt.grid(True)
# Set the x-axis and y-axis labels
plt.xlabel('Profit')
plt.ylabel('Density')

# Add a title to the plot
plt.title('Profit Distribution')

# Display the plot
plt.show()
```

```python
# Set the dark background style
plt.style.use('seaborn')
# Create the kernel density estimate plot with seaborn
sns.kdeplot(data=df['Quantity'], fill=True, color='blue', alpha=0.3, linewidth=2)

# Add a vertical line for the median
plt.axvline(df['Quantity'].median(), color='red', linestyle='--', linewidth=2)
plt.grid(True)
# Set the x-axis and y-axis labels
plt.xlabel('Quantity')
plt.ylabel('Density')

# Add a title to the plot
plt.title('Quantity Distribution')

# Display the plot
plt.show()
```

```python
# Create the kernel density estimate plot with seaborn
sns.kdeplot(data=df['Discount'], fill=True, color='blue', alpha=0.3, linewidth=2)

# Add a vertical line for the median
plt.axvline(df['Discount'].median(), color='red', linestyle='--', linewidth=2)

plt.style.use('default')
plt.grid(True)

# Set the x-axis and y-axis labels
plt.xlabel('Discount')
plt.ylabel('Density')

# Add a title to the plot
plt.title('Discount Distribution')

# Display the plot
plt.show()
```

### Multivariate Analysis :

```python
# Group the data and calculate the sum for each category
grouped_data = df.groupby('Category')['Sales', 'Quantity', 'Discount', 'Profit'].sum()

# Create the bar plot
grouped_data.plot.bar()
#set style
plt.style.use('default')
# Set the x-axis label
plt.xlabel('Category')

# Set the y-axis label
plt.ylabel('Value')

# Set the plot title
plt.title('Sum of Sales, Quantity, Discount, and Profit per Category')

# Display the plot
plt.show();
```

#### The above graph shows that the sales is highest for Category 'Technology'.

```python
# Set the figure size
plt.figure(figsize=(20, 15))

# Group the data and calculate the sum for each sub-category
grouped_data = df.groupby('Sub-Category')['Sales', 'Quantity', 'Discount', 'Profit'].sum()

# Create the bar plot
grouped_data.plot.bar()
#set style
plt.style.use('default')
# Set the x-axis label
plt.xlabel('Sub-Category')

# Set the y-axis label
plt.ylabel('Value')

# Set the plot title
plt.title('Sum of Sales, Quantity, Discount, and Profit per Sub-Category')

# Display the plot
plt.show();
```

#### The above graph shows that the sales is highest for Sub-Category 'Phones'.

### Heatmap :

```python
# Compute the correlation matrix
corr_matrix = df[['Sales', 'Quantity', 'Discount', 'Profit']].corr()

# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap="Reds")

# Set the plot title
plt.title('Correlation Heatmap')

# Display the plot
plt.show()
```

## The Analysis' results

The Category 'Office supplies' has the maximum quantity sold and discount applied, indicating a high demand for these products. On the other hand, the Category 'Technology' has the lowest quantity sold, suggesting potential opportunities for improvement. However, despite lower quantity, the Category 'Technology' generates the highest profit, making it a focal point for maximizing returns.

Certain Sub-Categories, namely 'Bookcases', 'Supplies', and 'Tables', result in losses. These Sub-Categories require attention and further analysis to identify areas of improvement and potential cost-saving measures.

The State 'Wyoming' stands out for achieving the highest sales, indicating a strong market presence and potential growth opportunities in that region.

The Category 'Office supplies' not only records the maximum sales but also has the highest count in the 'Consumer' segment. This suggests a significant consumer demand for office supplies, highlighting the importance of catering to the needs of this segment to further enhance sales and profitability.

By focusing on these findings, businesses can optimize their strategies, prioritize investments, and allocate resources effectively to drive growth, minimize losses, and capitalize on profitable opportunities.
